```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Assuming your original DataFrame is named 'data'
data = pd.read_csv("/content/02-14-2018.csv")

# Selecting the first 10 columns as features
x = data.iloc[:, :10]

# Selecting the last column ('Label') as the target variable
y = data['Label']

# Label encoding for categorical target variable if needed
enc = LabelEncoder()
y = enc.fit_transform(y)

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Load your dataset and perform any necessary preprocessing
data = pd.read_csv("/content/02-14-2018.csv")

# Assuming your original DataFrame is named 'data'
# Selecting the first 10 columns as features
a = data.iloc[:, :10]

# Assuming 'Timestamp' is the name of the column
a['Timestamp'] = pd.to_datetime(a['Timestamp'], format='%d/%m/%Y %H:%M:%S')
a['Timestamp'] = a['Timestamp'].astype(int) / 10**9  # Convert to Unix timestamp
# Drop 'Timestamp' column
a = a.drop(columns=['Timestamp'])

# Selecting the last column ('Label') as the target variable
b = data['Label']

# Label encoding for categorical target variable if needed
enc = LabelEncoder()
b = enc.fit_transform(b)

# Split the dataset into training and testing sets
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.3, random_state=10)

# Create a CatBoost classifier
catboost_model = CatBoostClassifier()

# Define the hyperparameters to search
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'depth': [4, 6, 8],
    'iterations': [100, 200, 300]
}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(catboost_model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_search.fit(a_train, b_train)

# Get the best parameters and train the final model
best_params = grid_search.best_params_
final_model = CatBoostClassifier(**best_params)
final_model.fit(a_train, b_train)

# Make predictions on the test set
b_pred = final_model.predict(a_test)

# Inverse transform label encoding if needed
b_test_original = enc.inverse_transform(b_test)
b_pred_original = enc.inverse_transform(b_pred)

# Evaluate the accuracy
accuracy = accuracy_score(b_test, b_pred)
print(f"Accuracy: {accuracy}")
```
